```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set Matplotlib Parameters
plt.style.use('fivethirtyeight')
plt.rc('text', usetex=True)   #  Might require installing TeX fonts
plt.rc('axes', facecolor='w', labelcolor='b', edgecolor='k', labelsize=18)
plt.rcParams['font.size'] = 16
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 16
```

```python
data = {
    "Model": ["GPT-4", "MEIRa-H", "Code Llama-34B"] * 2,
    "Macro-F1": [90.7, 88.5, 67, 88.8, 86.1, 68.9],
    "Dataset": ["FantasyCoref"] * 3 + ["LitBank"] * 3
}

link_df = pd.DataFrame(data)

datasets = link_df['Dataset'].unique()
models = link_df['Model'].unique()

colors_shade = ["#a6c8f1","#b4efb4", "#ffdfcc","#bfa0d5", '#e76a6a']

# plt.figure(figsize=(10, 6))
plt.figure(figsize=(12, 6))


# Define bar width and bar positions
bar_width = 0.2
index = np.arange(len(datasets))

# Plotting each model's bar in each dataset
for i, model in enumerate(models):
    model_data = link_df[link_df['Model'] == model]
    dataset_means = [model_data[model_data['Dataset'] == dataset]['Macro-F1'].values[0] for dataset in datasets]
    plt.bar(index + i * bar_width, dataset_means, bar_width, label=model, color=colors_shade[i], edgecolor='black', linewidth=1.5)
    ## Add value on top of the bar
    for j, value in enumerate(dataset_means):
        plt.text(index[j] + i * bar_width, value + 0.5, str(value), ha='center', va='bottom')

# Adding labels and title
plt.title('Linking Performance on Different Datasets')
plt.ylabel('Macro-F1 Score', color='grey')
plt.xticks(index + bar_width * (len(models) / 2 - 0.5), datasets)

# Add a grid for the y-axis
plt.grid(True, axis='both', linestyle='--', alpha=0.7, linewidth=0.7)

# Set minimum and maximum values for y-axis
plt.ylim(0, 100)

# Adding a legend
plt.legend(title='Model')

# ## legend upper right
# plt.legend(title='Model', loc='upper right')

## legend down outside
# plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.15), ncol=3)
plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1), )

# Adjust layout for better fit
plt.tight_layout()

# Display the plot
plt.show()
```
